# Cuaderno 16b: Dibujando toures del TSP

```python
import matplotlib.pyplot as plt
import random

def dibujarTour(coordx, coordy, tour):
    plt.plot(coordx, coordy, 'ro')
    Tx = [coordx[i] for i in tour]
    Ty = [coordy[i] for i in tour]
    Tx.append(coordx[tour[0]])
    Ty.append(coordy[tour[0]])
    plt.plot(Tx, Ty, 'b-')
    plt.show()
```

```python
# crear 30 coordenadas con numeros al azar entre 0 y 100
n = 30
x = [random.randint(0,10) for i in range(n)]
y = [random.randint(0,10) for i in range(n)]

# crear tour con ordenamiento trivial
T = range(n)
```

```python
# dibujar Tour
dibujarTour(x, y, T)
```
